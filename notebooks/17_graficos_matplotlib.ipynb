# Gráficos en Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Tipos de Gráficos

```python
# Datos para los gráficos
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
data = np.random.randn(1000)
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
```

### 1. Gráficos de Líneas (Line plots)

- Definición: Representan datos en una línea continua a lo largo de un eje.
- Uso: Ideal para visualizar tendencias a lo largo del tiempo o la relación entre dos variables.

```python
# 1. Gráfico de Líneas
plt.figure(figsize=(10, 3))
plt.plot(x, y, label='Seno')
plt.plot(x, z, label='Coseno')
plt.title('Gráfico de Líneas')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

### 2. Gráficos de Barras (Bar plots)

- Definición: Utilizan barras rectangulares para representar datos.
- Uso: Útiles para comparar cantidades entre diferentes categorías.

```python
# 2. Gráfico de Barras
plt.figure(figsize=(10, 3))
plt.bar(categories, values, color='skyblue')
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

### 3. Gráficos de Barras Apiladas (Stacked bar plots)

- Definición: Barras divididas en secciones apiladas una sobre otra.
- Uso: Para mostrar la composición de diferentes categorías y su contribución al total.

```python
# 3. Gráfico de Barras Apiladas
values2 = [3, 6, 2, 7]
plt.figure(figsize=(10, 3))
plt.bar(categories, values, color='skyblue', label='Valores 1')
plt.bar(categories, values2, bottom=values, color='lightgreen', label='Valores 2')
plt.title('Gráfico de Barras Apiladas')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.legend()
plt.show()
```

### 4. Histogramas (Histograms)

- Definición: Muestran la distribución de un conjunto de datos continuos.
- Uso: Para visualizar la frecuencia de valores en intervalos específicos.

```python
# 4. Histograma
plt.figure(figsize=(10, 3))
plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

### 5. Gráficos de Dispersión (Scatter plots)

- Definición: Utilizan puntos para representar la relación entre dos variables.
- Uso: Para identificar patrones, correlaciones o tendencias entre dos variables.

```python
# 5. Gráfico de Dispersión
plt.figure(figsize=(10, 3))
plt.scatter(x, y, color='red')
plt.title('Gráfico de Dispersión')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### 6. Gráficos de Pastel (Pie charts)

- Definición: Representan datos en forma de "rebanadas" de un círculo.
- Uso: Para mostrar la proporción de diferentes categorías en un todo.

```python
# 6. Gráfico de Pastel
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Gráfico de Pastel')
plt.show()
```

### 7. Gráficos de Área (Area plots)

- Definición: Similares a los gráficos de líneas pero con el área debajo de la línea rellena.
- Uso: Para mostrar la evolución de cantidades a lo largo del tiempo y su contribución total.

```python
# 7. Gráfico de Área
plt.figure(figsize=(10, 6))
plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.plot(x, y, color="Slateblue", alpha=0.6)
plt.title('Gráfico de Área')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### 8. Gráficos de Cajas y Bigotes (Box plots)

- Definición: Muestran la distribución de datos a través de sus cuartiles.
- Uso: Para identificar outliers y entender la dispersión y forma de la distribución de los datos.

```python
# 8. Gráfico de Cajas y Bigotes
data = np.random.rand(10, 4)
plt.figure(figsize=(10, 6))
plt.boxplot(data, vert=True, patch_artist=True, labels=categories)
plt.title('Gráfico de Cajas y Bigotes')
plt.show()
```

### 9. Gráficos de Violín (Violin plots)

- Definición: Combina características de los box plots y los gráficos de densidad.
- Uso: Para comparar la distribución de varios grupos.

```python
# 9. Gráfico de Violín
plt.figure(figsize=(10, 6))
plt.violinplot(data, showmeans=False, showmedians=True)
plt.title('Gráfico de Violín')
plt.show()
```

### 10. Gráficos de Tallo (Stem plots)

- Definición: Utilizan líneas verticales para representar datos.
- Uso: Para datos discretos y comparar valores individuales.

```python
# 10. Gráfico de Tallo
markerline, stemlines, baseline = plt.stem(x, y, '-.')
plt.setp(markerline, 'markerfacecolor', 'b')
plt.title('Gráfico de Tallo')
plt.show()
```

### 11. Gráficos de Paso (Step plots)

- Definición: Muestran cambios abruptos en valores en intervalos discretos.
- Uso: Para datos que cambian en pasos en lugar de continuamente.

```python
# 11. Gráfico de Paso
plt.figure(figsize=(10, 6))
plt.step(x, y, label='Seno')
plt.step(x, z, label='Coseno', linestyle='--')
plt.title('Gráfico de Paso')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

### 12. Gráficos de Contorno (Contour plots)

- Definición: Representan datos tridimensionales en dos dimensiones usando contornos.
- Uso: Para mostrar niveles de valores de una función de dos variables.

```python
# 12. Gráfico de Contorno
X, Y = np.meshgrid(x, x)
Z = np.sin(X) * np.cos(Y)
plt.figure(figsize=(10, 6))
cp = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(cp)
plt.title('Gráfico de Contorno')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### 13. Gráficos de Imagen (Image plots)

- Definición: Muestran datos en una cuadrícula donde los valores se representan con colores.
- Uso: Para datos matriciales y visualización de imágenes.

```python
# 13. Gráfico de Imagen
img = np.random.rand(10, 10)
plt.figure(figsize=(8, 8))
plt.imshow(img, cmap='hot', interpolation='nearest')
plt.title('Gráfico de Imagen')
plt.show()
```

### 14. Gráficos de Matrices de Correlación (Correlation matrix plots)

- Definición: Muestran la relación entre múltiples variables.
- Uso: Para identificar correlaciones entre variables en un conjunto de datos.

```python
# 14. Gráficos de Matrices de Correlación
df = pd.DataFrame(np.random.rand(10, 10), columns=list('ABCDEFGHIJ'))
corr = df.corr()
plt.figure(figsize=(10, 8))
plt.matshow(corr, fignum=1)
plt.colorbar()
plt.title('Gráficos de Matrices de Correlación')
plt.show()
```

### 15. Gráficos 3D (3D plots)

- Definición: Visualizan datos en tres dimensiones.
- Uso: Para representar relaciones complejas entre tres variables.

```python
# 15. Gráficos 3D
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='Seno-Coseno')
ax.set_title('Gráfico 3D')
plt.show()
```

### 16. Gráficos de Errores (Error bar plots)

- Definición: Muestran los valores con barras de error.
- Uso: Para representar la incertidumbre en los datos.

```python
# 16. Gráfico de Errores
x = np.linspace(0, 10, 10)
y = np.sin(x)
yerr = np.random.rand(10) * 0.1
plt.figure(figsize=(10, 6))
plt.errorbar(x, y, yerr=yerr, fmt='-o')
plt.title('Gráfico de Errores')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### 17. Gráficos de Radar (Radar charts)

- Definición: Muestran datos multivariables en un formato circular.
- Uso: Para comparar múltiples variables a la vez.

```python
# 17. Gráfico de Radar
labels = np.array(['A', 'B', 'C', 'D'])
stats = np.array([4, 3, 2, 5])
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
stats = np.concatenate((stats,[stats[0]]))
angles += angles[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, stats, color='red', alpha=0.25)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title('Gráfico de Radar')
plt.show()
```

### 18. Subgráficos (Subplots)

- Definición: Permiten colocar múltiples gráficos en una sola figura.
- Uso: Para comparar diferentes conjuntos de datos o diferentes aspectos del mismo conjunto de datos.

```python
# 18. Subgráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].plot(x, y)
axs[0, 0].set_title('Seno de x')

axs[0, 1].plot(x, np.cos(x), 'tab:orange')
axs[0, 1].set_title('Coseno de x')

axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('-Seno de x')

axs[1, 1].plot(x, -np.cos(x), 'tab:red')
axs[1, 1].set_title('-Coseno de x')

fig.tight_layout()
plt.show()
```
