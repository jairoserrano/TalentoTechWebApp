```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Tipos de Gráficos

```python
# Datos de ejemplo
df = pd.DataFrame({
    'día': ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'],
    'ventas': [20, 35, 30, 35, 27, 25, 32],
    'gastos': [22, 34, 29, 33, 26, 23, 30]
})
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
```

### 1. Gráficos de Barras (Bar plots)

- Definición: Barras rectangulares para representar datos categóricos.
- Uso: Para comparar cantidades entre diferentes categorías.

```python
# 1. Gráfico de Barras
plt.figure(figsize=(10, 3))
sns.barplot(x='día', y='ventas', data=df)
plt.title('Gráfico de Barras')
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.show()
```

### 2. Gráficos de Líneas (Line plots)

- Definición: Representan datos en una línea continua.
- Uso: Para visualizar tendencias a lo largo del tiempo.

```python
# 2. Gráfico de Líneas
plt.figure(figsize=(10, 3))
sns.lineplot(x='día', y='ventas', data=df, label='Ventas')
sns.lineplot(x='día', y='gastos', data=df, label='Gastos')
plt.title('Gráfico de Líneas')
plt.xlabel('Día')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```

### 3. Gráficos de Dispersión (Scatter plots)

- Definición: Utilizan puntos para representar la relación entre dos variables.
- Uso: Para identificar patrones y correlaciones entre dos variables.

```python
# 3. Gráfico de Dispersión
plt.figure(figsize=(10, 3))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='time', style='time', palette='deep')
plt.title('Gráfico de Dispersión')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')
plt.show()
```

### 4. Gráficos de Regresión (Regression plots)

- Definición: Muestran la relación entre dos variables con una línea de regresión ajustada.
- Uso: Para visualizar la tendencia y la relación entre variables.

```python
# 4. Gráfico de Regresión
plt.figure(figsize=(10, 3))
sns.regplot(x='total_bill', y='tip', data=tips)
plt.title('Gráfico de Regresión')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')
plt.show()
```

### 5. Gráficos de Cajas y Bigotes (Box plots)

- Definición: Muestran la distribución de datos a través de sus cuartiles.
- Uso: Para identificar outliers y entender la dispersión de los datos.

```python
# 5. Gráfico de Cajas y Bigotes
plt.figure(figsize=(10, 3))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Gráfico de Cajas y Bigotes')
plt.xlabel('Día')
plt.ylabel('Total de la Cuenta')
plt.show()
```

### 6. Gráficos de Violín (Violin plots)

- Definición: Combinan características de los box plots y los gráficos de densidad.
- Uso: Para comparar la distribución de varios grupos.

```python
# 6. Gráfico de Violín
plt.figure(figsize=(10, 3))
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Gráfico de Violín')
plt.xlabel('Día')
plt.ylabel('Total de la Cuenta')
plt.show()
```

### 7. Gráficos de Enjambre (Swarm plots)

- Definición: Muestran todos los puntos de datos en una gráfica categórica sin solaparlos.
- Uso: Para visualizar la distribución y la densidad de datos categóricos.

```python
# 7. Gráfico de Enjambre
plt.figure(figsize=(10, 3))
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.title('Gráfico de Enjambre')
plt.xlabel('Día')
plt.ylabel('Total de la Cuenta')
plt.show()
```

### 8. Gráficos de Puntos (Point plots)

- Definición: Muestran los valores medios de datos categóricos con barras de error.
- Uso: Para comparar las medias de diferentes categorías.

```python
# 8. Gráfico de Puntos
plt.figure(figsize=(10, 3))
sns.pointplot(x='day', y='total_bill', data=tips)
plt.title('Gráfico de Puntos')
plt.xlabel('Día')
plt.ylabel('Total de la Cuenta')
plt.show()
```

### 9. Gráficos de Mapas de Calor (Heatmaps)

- Definición: Representan datos matriciales con colores.
- Uso: Para visualizar la densidad o intensidad de datos en una matriz.

```python
# 9. Mapas de Calor
plt.figure(figsize=(12, 4))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Mapa de Calor')
plt.show()
```

### 10. Gráficos de Pares (Pair plots)

- Definición: Muestran relaciones entre pares de variables en un conjunto de datos.
- Uso: Para explorar relaciones entre múltiples variables.

```python
# 10. Gráficos de Pares
plt.figure(figsize=(2, 2))
sns.pairplot(tips)
plt.title('Gráficos de Pares')
plt.show()
```

### 11. Gráficos de Distribución (Distribution plots)

- Definición: Muestran la distribución de un conjunto de datos.
- Uso: Para visualizar la forma, centralidad y dispersión de datos.

```python
# 11. Gráficos de Distribución
plt.figure(figsize=(10, 3))
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Gráfico de Distribución')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Frecuencia')
plt.show()
```

### 12. Gráficos de Join (Joint plots)

- Definición: Combinan gráficos de dispersión y gráficos de distribución.
- Uso: Para estudiar la relación entre dos variables junto con sus distribuciones marginales.

```python
# 12. Gráficos de Join
plt.figure(figsize=(10, 3))
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
plt.title('Gráfico de Join')
plt.show()
```

### 13. Gráficos de Facetas (Facet grids)

- Definición: Dividen los datos en subconjuntos y muestran un gráfico para cada subconjunto.
- Uso: Para comparar múltiples subconjuntos de datos categóricos.

```python
# 13. Gráficos de Facetas
g = sns.FacetGrid(tips, col="time")
g.map(sns.histplot, "total_bill")
plt.show()
```

### 14. Gráficos de Boxen (Boxen plots)

- Definición: Variación de los box plots para grandes conjuntos de datos.
- Uso: Para visualizar la distribución y detectar outliers en grandes conjuntos de datos.

```python
# 14. Gráfico de Boxen
plt.figure(figsize=(10, 3))
sns.boxenplot(x='day', y='total_bill', data=tips)
plt.title('Gráfico de Boxen')
plt.xlabel('Día')
plt.ylabel('Total de la Cuenta')
plt.show()
```

### 15. Gráficos de Strip (Strip plots)

- Definición: Muestran todos los puntos de datos en una gráfica categórica.
- Uso: Para visualizar la distribución y la dispersión de datos categóricos.

```python
# 15. Gráfico de Strip
plt.figure(figsize=(10, 3))
sns.stripplot(x='day', y='total_bill', data=tips)
plt.title('Gráfico de Strip')
plt.xlabel('Día')
plt.ylabel('Total de la Cuenta')
plt.show()
```

### 16. Gráficos de Catplot (Catplot, para gráficos categóricos combinados)

- Definición: Interfaz unificada para varios tipos de gráficos categóricos.
- Uso: Para comparar y visualizar datos categóricos utilizando diferentes tipos de gráficos.

```python
# 16. Gráficos de Catplot
plt.figure(figsize=(10, 3))
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
plt.title('Gráfico de Catplot')
plt.xlabel('Día')
plt.ylabel('Total de la Cuenta')
plt.show()
```

### 17. Gráficos de Lineplot con Error (Lineplot with error bands)

- Definición: Gráficos de líneas con bandas de error para representar la incertidumbre.
- Uso: Para visualizar tendencias junto con la variabilidad o incertidumbre en los datos.

```python
# 17. Gráfico de Lineplot con Error
plt.figure(figsize=(10, 3))
sns.lineplot(x='día', y='ventas', data=df, label='Ventas')
sns.lineplot(x='día', y='gastos', data=df, label='Gastos')
plt.fill_between(df['día'], df['ventas']-3, df['ventas']+3, alpha=0.3)
plt.title('Gráfico de Lineplot con Error')
plt.xlabel('Día')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```
