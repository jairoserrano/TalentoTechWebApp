## Instrucciones iniciales

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generar datos de ejemplo
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
x = np.linspace(0, 10, 80)  # 100 puntos entre 0 y 10 
y = np.sin(x)  # Función seno
```

```python
# Crear el gráfico básico
plt.plot(x, y)
plt.show()
```

```python
# Generar datos de ejemplo
x = np.linspace(0, 10, 100)  # 100 puntos entre 0 y 10 
y = np.cos(x)  # Función seno
```

```python
# Crear el gráfico básico
plt.plot(x, y)
plt.show()
```

```python
# Añadir título y etiquetas
plt.plot(x, y)
plt.title('Gráfico de Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
# Ajustar el tamaño de la figura
plt.figure(figsize=(10, 4))  # Ancho de 10 pulgadas y alto de 6 pulgadas
plt.plot(x, y)
plt.title('Gráfico de Seno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
# Añadir una leyenda
plt.figure(figsize=(10, 3))
plt.plot(x, y, label='Coseno')
plt.title('Gráfico de Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()  # Mostrar la leyenda
plt.show()
```

```python
# Personalización de líneas y colores
plt.figure(figsize=(10, 3))
plt.plot(x, y, 'r--')  # Línea roja discontinua
plt.title('Gráfico de Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
# Añadir una cuadrícula
plt.figure(figsize=(10, 3))
plt.plot(x, y)
plt.title('Gráfico de Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
```

```python
# Establecer límites en los ejes
plt.figure(figsize=(10, 3))
plt.plot(x, y)
plt.title('Gráfico de Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.grid(True)
plt.show()
```

```python
# Añadir anotaciones
plt.figure(figsize=(10, 3))
plt.plot(x, y)
plt.title('Gráfico de Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.annotate('Máximo', xy=(0.2, 1), xytext=(2, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
# Guardar el gráfico
plt.figure(figsize=(10, 3))
plt.plot(x, y)
plt.title('Gráfico de Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.savefig('grafico_coseno.pdf')
plt.show()
```

```python
# Ejemplo completo con todas las opciones
plt.figure(figsize=(10, 3))
plt.plot(x, y, 'r--', label='Coseno')  # Línea roja discontinua
plt.title('Gráfico de Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid(True)
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.annotate('Línea', xy=(1.40, 0.25), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.savefig('grafico_seno_completo.pdf')
plt.show()
```
